```python
from mtcnn import MTCNN
import cv2
```

```python
!ls
```

```python
img = cv2.imread("face1.jpeg")
detector = MTCNN()
print(detector.detect_faces(img))
```

```python
faces = detector.detect_faces(img)
for face in faces:
  x = face['box'][0]
  y = face['box'][1]
  width = face['box'][2]
  height = face['box'][3]
  cv2.rectangle(img, (x,y),(x+width, y+height),(0,0,255),3)
```
